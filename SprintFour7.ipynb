```python
#facial detector using webcam to detect eye and face and smile.

#import needed lib
import cv2
import numpy as np


# create classifier object for face detection
face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

smile = cv2.CascadeClassifier('haarcascade_smile.xml')

#create eye classifier object for eye detection
eye = cv2.CascadeClassifier('haarcascade_eye.xml')

#capture images using my webcam
vid = cv2.VideoCapture(0)

# create loop to iterate through frames.
while 1:
    #capture first frame from webcam
    ret, image_1 = vid.read()
    
    #read image as gray scale
    gray_image = cv2.cvtColor(image_1, cv2.COLOR_BGR2GRAY)
    
    #to detect face
    faces = face.detectMultiScale(gray_image, 1.3, 5)
    
    #for drawing rectangle around the face
    for (x,y,w,h) in faces:
        cv2.rectangle(image_1,(x,y),(x+w,y+h),(255,0,0),2)
        eye_gray = gray_image[y:y+h, x:x+w]
        eye_color = image_1[y:y+h, x:x+w]
       
        
        #to detect eyes
        eye_detect = eye.detectMultiScale(eye_gray)
        
        #for drawing rectangle around the eye
        for (ex,ey,ew,eh) in eye_detect:
            cv2.rectangle(eye_color,(ex,ey),(ex+ew,ey+eh),(0,0,255),2)
        
        #same as eye but changing color and mineighbor
        smile_detect = smile.detectMultiScale(eye_gray, 1.5, 18)
        
        #rectangle around smile
        for (sx, sy, sw, sh) in smile_detect:
            cv2.rectangle(eye_color,(sx,sy),(sx+sw,sy+sh),(0,255,0),2)
            
    # to show image/video
    cv2.imshow('image1',image_1)
    
    #for when user presses escape key to exit program
    pressKey = cv2.waitKey(30) & 0xff
    if pressKey == 27:
        break

# release frame in milliseconds
vid.release()

#end program
cv2.destroyAllWindows()
```
